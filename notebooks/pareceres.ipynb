```python
from bs4 import BeautifulSoup
import json
import re
```

```python
with open("dados/parecer_final.json") as fl:
    text = ''
    line = fl.readline()
    while line:
        text += line.replace('\n', '') + ' '
        line = fl.readline()
    
    jsn_pareceres = json.loads(text)
    parecer_html = BeautifulSoup(jsn_pareceres['X-TIKA:content'], 'html.parser')
    
parecer =parecer_html.get_text()
```
